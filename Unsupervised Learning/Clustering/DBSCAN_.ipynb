```python
sc = StandardScaler()

x = sc.fit_transform(df)
x
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN()
db.fit(x)
```

```python
db.labels_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,db.labels_)
```

```python
db  =DBSCAN(eps=2, min_samples=3)

db.fit_predict(x)
```

```python
silhouette_score(x,db.labels_)
```

```python
df['Y'] = db.labels_
```
