```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('Mall_Data.csv', index_col=0)
df
```

```python
sc = StandardScaler()

x = sc.fit_transform(df)
x
```

```python
db = DBSCAN()
y = db.fit_predict(x)
y
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,y)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=4).fit(x)
dist,ind = neigh.kneighbors(x)
```

```python
dist = np.sort(dist, axis=0)
dist
```

```python
d = dist[:,1]
d
```

```python
plt.plot(d)
plt.xlabel('Number of samples')
plt.ylabel('Epsilon Value')
plt.title('K-dist Plot')
```

```python
db = DBSCAN(eps=0.35,min_samples=3)
y = db.fit_predict(x)
y
```

```python
silhouette_score(x,y)
```

```python
df['y'] = y
```

```python
df['y'].value_counts()
```

```python
sns.scatterplot(x[:,0],x[:,1], hue = df['y'], palette='gnuplot')
```

- DBSCAN is able to find the outliers.
- All the outliers are given label of -1.
