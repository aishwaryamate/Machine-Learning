```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Iris.csv', index_col=0)
df
```

```python
x = df.iloc[:,:-1]
y = df['Species']
```

```python
le = LabelEncoder()

y = le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
dt = DecisionTreeClassifier(criterion='entropy')
dt.fit(xtrain,ytrain)

ypred = dt.predict(xtest)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred))
```

```python
print(f"Training accuracy is: {dt.score(xtrain,ytrain)}")
print(f"Testing accuracy is : {dt.score(xtest,ytest)}")
```

# Regression

```python
df
```

```python
x = df.iloc[:,0:3]
y = df['PetalWidthCm']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(ytest,ypred)
```

```python
print(f"Training Accuracy: {dt.score(xtrain,ytrain)}")
print(f"Testing Accuracy: {dt.score(xtest,ytest)}")
```

# Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
DecisionTreeRegressor()
```

```python
params = {
    'criterion' : ["squared_error", "friedman_mse", "absolute_error"],
    'splitter' : ["random","best"],
    'max_depth' : [1,2,3,4,5,6]
}
```

```python
grid = GridSearchCV(DecisionTreeRegressor(), param_grid=params, verbose=2)
```

```python
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_
```

```python
reg = DecisionTreeRegressor(criterion='squared_error', max_depth=5, splitter='best')
reg.fit(xtrain,ytrain)
ypred = reg.predict(xtest)
```

```python
print(f"Training Accuracy: {reg.score(xtrain,ytrain)}")
print(f"Testing Accuracy: {reg.score(xtest,ytest)}")
```
