```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/cars.csv')
df
```

```python
df.replace('?',np.nan,inplace=True)
```

```python
df.isna().sum()
```

```python
df['normalized-losses'].skew()
```

```python
df['horsepower'].skew()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='most_frequent')
```

```python
df.loc[:,['normalized-losses','horsepower']] = si.fit_transform(df.loc[:,['normalized-losses','horsepower']])
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
cat_cols = df.select_dtypes(object).columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()
```

```python
df[cat_cols] = oe.fit_transform(df[cat_cols])
```

```python
df
```

```python
x = df.iloc[:,:-1]
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
dt.score(xtrain,ytrain)
```

```python
dt.score(xtest,ytest)
```

```python
dt.feature_importances_
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
DecisionTreeRegressor()
```

```python
params = {
    'criterion' : ["squared_error", "friedman_mse", "absolute_error"],
    'max_depth' : [1,2,3,4,5,6],
    'min_samples_split' : [4,6,8,10,12],
    'max_leaf_nodes' : [8,10,12,14,16,18],
    'min_samples_leaf' : [1,2,3,4,5]
}
```

```python
grid = GridSearchCV(DecisionTreeRegressor(), param_grid=params, cv = 5, verbose=2)
```

```python
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_
```

```python
reg = DecisionTreeRegressor(criterion='absolute_error',max_depth=6,
                           max_leaf_nodes=18,min_samples_split=4, min_samples_leaf=1)
```

```python
reg.fit(xtrain,ytrain)
ypred = reg.predict(xtest)
```

```python
reg.score(xtrain,ytrain),reg.score(xtest,ytest)
```
