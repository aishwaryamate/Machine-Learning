```python
# Import Libraries
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, classification_report
import matplotlib.pyplot as plt
from sklearn import tree
import seaborn as sns
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 1000
sns.set_theme(style='darkgrid', palette='rainbow')
import warnings
warnings.filterwarnings('ignore')
```

# Regression

```python
#Read data
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Boston.csv'
                ,index_col=0)
df
```

```python
df.isna().sum()
```

# Model building

```python
x = df.iloc[:,:-1]
y = df['medv']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
#Evaluating regressor
```

```python
xtest
```

```python
rf = RandomForestRegressor()
rf.fit(xtrain,ytrain)
ypred = rf.predict(xtest)
ypred
```

```python
r2_score(ytest,ypred)
```

```python
print('R2 value :', r2_score(ytest, ypred))
```

# Classification

```python
#Read the dataset
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Mobile_price_range.csv')
df
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
sns.countplot(x = df['price_range'])
```

```python
df['price_range'].replace([0,1,2,3],['low','medium','high','very high'], inplace=True)
```

```python
df
```

```python
#Evaluate model
```

```python
x = df.iloc[:,:-1]
y = df['price_range']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
rfc = RandomForestClassifier()
rfc.fit(xtrain,ytrain)
yp = rfc.predict(xtest)
```

```python
xtest
```

```python
yp
```

```python
print(classification_report(ytest,yp))
```
